**<u>AIM :</u> Perform linear and circular convolution of**
            - Sine and Square Signal(50% DC) of same frequency but opposite phase.
            - Sine and Square wave(25% DC) of frequency f(sine)=2*f(square).
	        - Cosine and Triangular wave of same frequency but opposite phase.

**<u>CODE with OUTPUT :</u>**

*Importing the libraries required*

```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
from scipy import signal
```

**User-defined function for performing linear convolution**

```python
def lconv(x,h):
    N1 = len(x)
    N2 = len(h)
    N = N1+N2-1
    y = np.zeros(N)


    m = N-N1
    n = N-N2

    #Padding zeros to x and h to make their length to N
    x =np.pad(x,(0,m),'constant')
    h =np.pad(h,(0,n),'constant')


    #Linear convolution
    for n in range (N):
        for k in range (N):
            y[n] = y[n]+x[n-k]*h[k]
    
    return y
```

**User-defined function for performing circular convolution**

```python
def circ_conv(x,h):
    lx = len(x)
    lh = len(h)
    maxm = max(lx,lh)
    #Padding zeros to x and h to make their length to N
    x =np.pad(x,(0,maxm-lx),'constant')
    h =np.pad(h,(0,maxm-lh),'constant')
    
    y=[]
    
    #circular shifting
    for i in range(maxm):
        col=[]
        for j in range(i,-1,-1):
            col.append(x[j])
        for j in range(maxm-1,i,-1):
            col.append(x[j])
        y.append(col)
    
    #matrix multiplication
    res = y @ h
    
    return res
```

- **Sine and Square Signal(50% DC) of same frequency but opposite phase.**

```python
t = np.linspace(0,1,1000,endpoint = True)
```

Square wave generation

```python
h = signal.square(2*(np.pi)*5*t, duty = 0.5)
plt.plot(t, h)
plt.title('Square wave')
```

Sine wave generation

```python
x = np.sin(2*(-(np.pi))*5*t)
plt.plot(t,x)
```

**Linear Convolution**

```python
#performing linear convolution without using in-built library
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
ax1.stem(x)
ax1.set_title('x(n) plot')
ax2.stem(h)
ax2.set_title('h(n) plot')
ax3.stem(lconv(x,h))
ax3.set_title('Linearly convoluted plot')
```

**Circular Convolution**

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
ax1.stem(x)
ax1.set_title('x(n) plot')
ax2.stem(h)
ax2.set_title('h(n) plot')
ax3.stem(circ_conv(x,h))
ax3.set_title('Circular convoluted plot')
```

- **Sine and Square wave(25% DC) of frequency f(sine)=2*f(square).**

```python
t = np.linspace(0, 1, 500, endpoint=False)
fsine = 8
fsquare = 4
```

Sine wave generation

```python
x = np.sin(m.pi*fsine*t)
plt.plot(t,x,'r')
plt.title('Sine wave')
```

Square wave generation

```python
h=signal.square( np.pi * fsquare * t,duty=0.25)
plt.plot(t,h)
```

**Linear Convolution**

```python
#performing linear convolution without using in-built library
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
ax1.stem(x)
ax1.set_title('x(n) plot')
ax2.stem(h)
ax2.set_title('h(n) plot')
ax3.stem(lconv(x,h))
ax3.set_title('Linearly convoluted plot')
```

**Circular Convolution**

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
ax1.stem(x)
ax1.set_title('x(n) plot')
ax2.stem(h)
ax2.set_title('h(n) plot')
ax3.stem(circ_conv(x,h))
ax3.set_title('Circular convoluted plot')
```

- **Cosine and Triangular wave of same frequency but opposite phase.**

```python
t = np.linspace(0, 1, 500)
fs=5
```

Triangular wave generation

```python
#generating the triangular wave
h= signal.sawtooth(2 * np.pi * fs * t, 0.5)
plt.plot(t, h)
plt.title("traingular wave of phase 2pi")
```

Cosine wave generation

```python
#generating the cosine wave
x=np.cos(2*np.pi*fs*t)
plt.plot(t,x)
plt.title("cosine wave of opposite phase")
```

**Linear Convolution**

```python
#performing linear convolution without using in-built library
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
ax1.stem(x)
ax1.set_title('x(n) plot')
ax2.stem(h)
ax2.set_title('h(n) plot')
ax3.stem(lconv(x,h))
ax3.set_title('Linearly convoluted plot')
```

**Circular Convolution**

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
ax1.stem(x)
ax1.set_title('x(n) plot')
ax2.stem(h)
ax2.set_title('h(n) plot')
ax3.stem(circ_conv(x,h))
ax3.set_title('Circular convoluted plot')
```
