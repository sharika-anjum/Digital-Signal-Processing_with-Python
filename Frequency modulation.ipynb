```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
modulation_frequency = 12.0
carrier_frequency = 20.0
modulation_index = 1.0
```

```python
time = np.arange(10000.0) / 10000.0
m = np.sin(2.0 * np.pi *modulation_frequency * time) * modulation_index
carrier = np.sin(2.0 * np.pi *carrier_frequency * time)
product = np.zeros_like(m)
```

```python
for i, t in enumerate(time):
    product[i] = np.sin(2.0 * np.pi * carrier_frequency * t + m[i])
```

```python
fig,axs = plt.subplots(1,3,figsize=(20,5))

axs[0].plot(m)
axs[0].set_xlabel('Message signal')

axs[1].plot(carrier)
axs[1].set_xlabel('Carrier signal')

axs[2].plot(product)
axs[2].set_xlabel('Output signal')
```
