```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
import math as m
```

### Designing IIR filter (Butterworth filter)

```python
fs = 1000; #Sampling frequency
n = 4 #filter order
fc = 100 #cutoff frequency
```

```python
#calculating the normalized frequency

wc = 2*fc/fs
```

```python
#Filter desiging
[b,a]=sig.butter(n,wc) #by default it takes low pass filter
```

```python
#plotting the frequency response
[w,h]=sig.freqz(b,a,worN=1000)
```

```python
w=fs*w/(2*m.pi)
h_db = 20*np.log10(abs(h))
```

```python
plt.plot(w,h_db)
plt.xlabel("Frequency in Hz")
plt.ylabel("Magnitude in dB")
plt.title("Butterworth IIR Low pass filter")
plt.grid('on')
```

```python
#Filter desiging
[b,a]=sig.butter(n,wc,btype='highpass') #by default it takes low pass filter

#plotting the frequency response
[w,h]=sig.freqz(b,a,worN=2000)

w=fs*w/(2*m.pi)
h_db = 20*np.log10(abs(h))

plt.plot(w,h_db)
plt.xlabel("Frequency in Hz")
plt.ylabel("Magnitude in dB")
plt.title("Butterworth IIR High pass filter")
plt.grid('on')
```

### Designing FIR filter

```python
N = 20 #no.of coefficients
fc = 100
wc = 2*fc/fs #normalized frequency
t = sig.firwin(N,wc)
```

```python
#plotting the frequency response
[w,h]=sig.freqz(t,worN=2000) #no.of samples we want = worN
```

```python
#calculation for normalization
w=fs*w/(2*m.pi)
h_db = 20*np.log10(abs(h))
```

```python
plt.plot(w,h_db)
plt.xlabel("Frequency in Hz")
plt.ylabel("Magnitude in dB")
plt.title("FIR filter")
plt.grid('on')
```
