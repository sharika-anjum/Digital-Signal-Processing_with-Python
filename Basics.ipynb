**Plotting Multiple figures**

```python
x = np.linspace(0,3,101)
y = np.sin(2*m.pi*2*x)
z = np.cos(2*m.pi*2*x)
```

```python
plt.plot(x,y,'r--')
plt.plot(x,z,'b--')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sine wave")
plt.grid(True)
plt.legend(["Cos wave","Sine wave"],loc = 'upper right')
```

**Stem plot**

```python
plt.stem(x,y,'r--') #for colour
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sine wave")
plt.grid(True) #for making grid
```

**Impulse Signal**

```python
n = np.arange(-10,20)
l = np.size(n)
#Find index of array n where the value is zero
ind = np.where(n==0)
#create impulse array
imp = np.zeros(l)
imp[ind]=l
#Plotting
plt.stem(n,imp)
plt.title('Impulse Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid(True)
```

**Subplot**

```python
fig,axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,0].set_title('Sine plot')
axs[0,1].stem(x,y)
axs[0,1].set_title('Stem plot')
axs[1,0].step(x,y)
axs[1,0].set_title('Step plot')
axs[1,1].stem(n,imp)
axs[1,1].set_title('Impulse')
```

**Stem plot using step signal**

```python
n = np.arange(-10,20)
l = np.size(n)
#Find index of array n where the value is zero
ind = np.where(n>=0)
#create impulse array
imp = np.zeros(l)
imp[ind]=l
#Plotting
plt.stem(n,imp)
plt.title('Step Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid(True)
```

**Create and Plot**

- Sine wave of amplitude 5 and frequency 200
- Cos wave of amplitude 2 and frequency 100

*For both the plots find their stem plot as well*

***Show all the 4 plots using subplot***

```python
import math as m
```

```python
#sine plot

fs = 200
ts = np.linspace(0,1,100)
ys = 5 * np.sin(5*m.pi*fs*ts)
plt.plot(ts,ys)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sine wave")
```

```python
#cos plot

fc = 100
tc = np.linspace(0,1,100)
yc = 2*np.cos(2*m.pi*fc*tc)
plt.plot(tc,yc)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("cos wave")
```

```python
#Cos stem wave
plt.stem(tc,yc,'r--') #for colour
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Cos stem wave")
plt.grid(True) #for making grid
```

```python
#Sine stem wave
plt.stem(ts,ys,'r--') #for colour
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sine stem wave")
plt.grid(True) #for making grid
```

```python
fig,axs = plt.subplots(2,2,figsize=(10,10))
axs[0,0].plot(ts,ys,'r--')
axs[0,0].set_title('Sine plot')
axs[0,1].plot(tc,yc,'b--')
axs[0,1].set_title('Cos plot')
axs[1,0].stem(ts,ys,'g')
axs[1,0].set_title('Sine stem plot')
axs[1,1].stem(tc,yc,'y')
axs[1,1].set_title('Cos stem plot')
```

**Noisy Sine**

```python
#sine plot

ns = ys+ 0.5*np.random.rand(len(ts))
plt.plot(ts,ns)
plt.xlabel("time in secs")
plt.ylabel("Amplitude")
plt.title("Noisy sine")
```

```python
#cos plot

nc = yc+ 0.8*np.random.rand(len(ts))
plt.plot(tc,nc)
plt.xlabel("time in secs")
plt.ylabel("Amplitude")
plt.title("Noisy cos")
```

```python
#Noisy Cos stem wave
plt.stem(tc,nc,'r--') #for colour
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Noisy Cos stem wave")
plt.grid(True) #for making grid
```

```python
#Noisy sine stem wave
plt.stem(ts,ys,'r--') #for colour
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Noisy sine stem wave")
plt.grid(True) #for making grid
```

```python
fig,axs = plt.subplots(4,2,figsize=(10,10))
axs[0,0].plot(ts,ys,'r--')
axs[0,0].set_title('Sine plot')
axs[0,1].plot(tc,yc,'b--')
axs[0,1].set_title('Cos plot')
axs[1,0].stem(ts,ys,'g')
axs[1,0].set_title('Sine stem plot')
axs[1,1].stem(tc,yc,'y')
axs[1,1].set_title('Cos stem plot')

axs[2,0].plot(ts,ns,'r--')
axs[2,0].set_title('Noisy Sine plot')
axs[2,1].plot(tc,nc,'b--')
axs[2,1].set_title('Noisy Cos plot')
axs[3,0].stem(ts,ns,'g')
axs[3,0].set_title('Noisy Sine stem plot')
axs[3,1].stem(tc,nc,'y')
axs[3,1].set_title('Noisy Cos stem plot')
```
