# Importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plotting sine signal

```python
import math as mt
fs=200
tPeriod = 1/fs
timeTotal = 2*tPeriod
t=np.linspace(0,timeTotal,100,endpoint = True)
y=np.sin(2*mt.pi*fs*t)

plt.plot(t,y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
```

# Sampling

## **Method 1**

```python
#Define Sampling frequency according to Nyquist Criterion
FSampling = 600
Tsampling = 1/FSampling #sampling time
sampArray = np.arange(0,timeTotal,Tsampling)

r=round(len(t)/len(sampArray))
Samp = np.arange(0,len(t),r)
sigSampled = y[Samp]
xSampled = t[Samp]
```

- **Plot Sampled Signal**

```python
plt.plot(t,y)
plt.stem(xSampled,sigSampled,'red')
```

- **Plot reconstructed signal** (Not an ideal method, its an emulation)

```python
plt.plot(t,y)
plt.plot(xSampled,sigSampled,'red')
```

## **Method 2**

```python
sig=[]
xsamp=[]
for i in range(0,len(t),r):
    sig.append(y[i])
    xsamp.append(t[i])
```

```python
plt.plot(t,y)
plt.stem(xsamp,sig,'red')
```
