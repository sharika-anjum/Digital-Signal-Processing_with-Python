```python
#importing modules
import numpy as np
from matplotlib import pyplot as plt
import math as m
from scipy import signal
```

Our Pseudo code to perform linear convolution *(For the method without using in-built function)*

```python
def conv(x,h):
    N1 = len(x)
    N2 = len(h)
    N = N1+N2-1
    y = np.zeros(N)


    m = N-N1
    n = N-N2

    #Padding zeros to x and h to make their length to N
    x =np.pad(x,(0,m),'constant')
    h =np.pad(h,(0,n),'constant')


    #Linear convolution
    for n in range (N):
        for k in range (N):
            y[n] = y[n]+x[n-k]*h[k]
    
    return y
```

- **x(n) = [1,2,3] and h(n) = [1,1]**

```python
#defining the series
x = [1,2,3]
h = [1,1]
```

```python
#performing linear convolution using in-built library
res1 = np.convolve(x,h,'full')
res1
```

```python
#performing linear convolution without using in-built library
conv(x,h)
```

**Clearly both the result matched, hence my convolution is right.** 
*Plotting them:*

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
ax1.stem(x)
ax1.set_title('x(n) plot')
ax2.stem(h)
ax2.set_title('h(n) plot')
ax3.stem(conv(x,h))
ax3.set_title('Linearly convoluted plot')
```

- **Find Convolution of a Sine wave (length 60, amplitude 10, 2 Cycles) with a saw tooth wave (length 40, starting from 0 to 5 and 4 cycles)**

*Creating a Sine wave (length 60, amplitude 10, 2 Cycles) and plotting it*

```python
t = np.arange(0,60,0.1)
y = 10*np.sin(m.pi*0.0675*t)
plt.plot(t,y,'r')
```

*Creating a saw tooth wave (length 40, starting from 0 to 5 and 4 cycles) and plotting it*

```python
timePoints = np.linspace(0, 1,500)
timePoints=40*timePoints
plt.plot(timePoints, 2.5*signal.sawtooth(2 * np.pi * 0.1 * timePoints)+2.5)
```

```python
#defining the two series
x=10*np.sin(m.pi*0.0675*t)
h=2.5*signal.sawtooth(2 * np.pi * 0.1 * timePoints)+2.5
```

```python
#performing linear convolution using in-built library
res1 = np.convolve(x,h,'full')
res1
```

```python
#performing linear convolution without using in-built library
conv(x,h)
```

**Clearly both the result matched, hence my convolution is right.** 
*PLotting them:*

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
ax1.stem(x)
ax1.set_title('x(n) plot')
ax2.stem(h)
ax2.set_title('h(n) plot')
ax3.stem(conv(x,h))
ax3.set_title('Linearly convoluted plot')
```
