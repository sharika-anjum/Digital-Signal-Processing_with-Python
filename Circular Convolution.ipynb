```python
#importing modules

import numpy as np
import math as m
import matplotlib.pyplot as plt
```

**User-defined function**

```python
def circ_conv(x,h):
    lx = len(x)
    lh = len(h)
    maxm = max(lx,lh)
    #Padding zeros to x and h to make their length to N
    x =np.pad(x,(0,maxm-lx),'constant')
    h =np.pad(h,(0,maxm-lh),'constant')
    
    y=[]
    
    #circular shifting
    for i in range(maxm):
        col=[]
        for j in range(i,-1,-1):
            col.append(x[j])
        for j in range(maxm-1,i,-1):
            col.append(x[j])
        y.append(col)
    
    #matrix multiplication
    res = y @ h
    
    return res
```

- **x(n) = [2,1,2,1] and h(n) = [1,2,3,4]**

```python
x=[2,1,2,1]
h=[1,2,3,4]
```

```python
circ_conv(x,h)
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
ax1.stem(x)
ax1.set_title('x(n) plot')
ax2.stem(h)
ax2.set_title('h(n) plot')
ax3.stem(circ_conv(x,h))
ax3.set_title('Circularly convoluted plot')
```

- **Find Convolution of a Sine wave (length 60, amplitude 10, 2 Cycles) with a saw tooth wave (length 40, starting from 0 to 5 and 4 cycles)**

*Creating a Sine wave (length 60, amplitude 10, 2 Cycles) and plotting it*

```python
t = np.arange(0,60,0.1)
y = 10*np.sin(m.pi*0.0675*t)
plt.plot(t,y,'r')
```

*Creating a saw tooth wave (length 40, starting from 0 to 5 and 4 cycles) and plotting it*

```python
from scipy import signal
timePoints = np.linspace(0, 1,500)
timePoints=40*timePoints
plt.plot(timePoints, 2.5*signal.sawtooth(2 * np.pi * 0.1 * timePoints)+2.5)
```

```python
#defining the two series
x=10*np.sin(m.pi*0.0675*t)
h=2.5*signal.sawtooth(2 * np.pi * 0.1 * timePoints)+2.5
```

```python
circ_conv(x,h)
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))
ax1.stem(x)
ax1.set_title('x(n) plot')
ax2.stem(h)
ax2.set_title('h(n) plot')
ax3.stem(circ_conv(x,h))
ax3.set_title('Circularly convoluted plot')
```
